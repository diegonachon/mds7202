### 2.4 Modelos de ML [0.5 puntos]

Como corresponde a un problema de clasificación con clases altamente desbalanceadas, es necesario escoger un modelo/algoritmo que pueda manejar esto adecuadamente. Por lo anterior, se decide usar *Extreme Gradient Boosting* (XGBoost), *K-Nearest Neighbor* (KNN) y *Balanced Random Forest* (BRF).

Las principales ventajas de cada uno vienen dadas a continuación, además de los hiperparámetros más relevantes:

* XGBoost:
    
    * Ventajas:

        * **Manejo de desbalance**: Permite la modificación de los pesos de las clases, pudiendo ajustar los pesos de la clase minoritaria (casos fraudulentos) para que tengan más influencia durante el entrenamiento.

        * **Capacidad para capturar relaciones complejas** Como es un algoritmo basado en árboles, puede capturar relaciones no lineales y complejas entre las características.

        * **Regularización**: Permite regularizar segun L1 y L2, lo que ayuda a prevenir el sobreajuste.

        * **Eficiencia y velocidad**: Este algoritmo es eficiente en términos de tiempo de ejecución y memoria, por lo que puede ser ventajoso en caso de tener muchos datos.
        
    * Hiperparámetros relevantes:
    
        * **n_estimators**: Define el número de árboles en el modelo. Aumentar el número de árboles generalmente mejora la precisión del modelo, al reducir el sesgo, pero también aumenta el riesgo de sobreajuste y el tiempo de entrenamiento.
        
        * **learning_rate**: Controla la tasa a la que el modelo aprende. Un *learning rate* bajo puede producir un modelo más robusto y menos propenso al sobreajuste, pero requerirá más arboles para converger; mientras que un valor más alto puede acelerar el proceso de aprendizaje, pero aumentar el riesgo de sobreajuste.
        
        * **max_depth**: Es la profundidad máxima de cada árbol. Controla la complejidad del modelo, donde un valor alto permite capturar relaciones complejas en los datos, pero tambien aumentar el riesgo de sobreajuste; mientras que un valor bajo puede resultar en un modelo subajustado que no capta adecuadamente las relaciones en los datos.
        
        
* KNN:

    * Ventajas:
    
        * **Simplicidad e interpretabilidad**: Es un algoritmo intuitivo y fácil de entender, donde en cada iteración se buscan los k vecinos más cercanos en el conjunto de entrenamiento para asignar las clases.

        * **Adaptabilidad**: Puede ser adaptado para manejar el desbalance de clases, por medio de las distintas métricas de distancia y ponderación que posee.

        * **Identificación de zonas raras**: Con un k adecuado, se pueden identificar los datos anómalos, en este caso, los fraudulentos.
        
    * Hiperparámetros relevantes:
        
        * **n_neighbors**: El número de vecinos más cercanos a considerar para la clasificación. Un valor pequeño puede hacer que el modelo sea sensible al ruido (sobreajuste), mientras que un valor grande puede hacer que el modelo sea demasiado general (subajuste).
        
        * **weights**: Es un método de ponderación de las contribuciones de los vecinos, donde se define si según la distancia es el peso que se les da, o que todos pesen lo mismo sin importar esto. Esto puede influir significativamente en la precisión del modelo, donde pueden haber casos que los vecinos cercanos son más relevantes para la predicción.
        
        * **metric**: Corresponde a la métrica de distancia utilizada para encontrar los vecinos más cercanos. Va a depender según el tipo de *features* que se tengan.
        

* BRF:

    * Ventajas:
    
        * **Manejo explícito del desbalance**: Es un algoritmo variante de RF, que ajusta el problema del desbalance tomando submuestras balanceadas de las clases en la construcción de cada árbol. Eso hace que cada árbol en el bosque esté entrenado con una muestra equilibrada de las clases.

        * **Robustez y reducción de sobreajuste**: Es robusto al sobreajuste debido a su naturaleza de conjunto y al promediar los distintos árboles.

        * **Eficiencia en grandes conjuntos de datos**: Es eficiente y escalable, manejando bien grandes volúmenes de datos.
        
    * Hiperparámetros relevantes:
    
        * **n_estimators**: Define el número de árboles en el modelo. Aumentar el número de árboles generalmente mejora la precisión del modelo, al reducir el sesgo, pero también aumenta el riesgo de sobreajuste y el tiempo de entrenamiento.
        
        * **max_depth**: Es la profundidad máxima de cada árbol. Controla la complejidad del modelo, donde un valor alto permite capturar relaciones complejas en los datos, pero tambien aumentar el riesgo de sobreajuste; mientras que un valor bajo puede resultar en un modelo subajustado que no capta adecuadamente las relaciones en los datos.
        
        * **sampling_strategy**: Corresponde a la estratgia de muestreo utilizada para equilibrar las clases. Según el tipo utilizado, se asegura que cada árbol en el bosque se entrene con un conjunto de datos balanceado, mejorando la capacidad del modelo para detectar la clase minoritaria.

#### 2.5.1 Optimización de hiperparámetros

Para la optimización del modelo se consideran los siguientes hiperparámetros, teniendo como función objetivo la maximización del *f1-score*, y un total de 50 trials.

* **n_estimators**: Define el número de árboles en el bosque. Un mayor número de árboles generalmente mejora el rendimiento del modelo al reducir la varianza, pero incrementa el costo computacional.

* **max_depth**: Corresponde a la profundidad máxima de lo árboles, que controla cuán complejas son las interacciones que el modelo puede capturar. Una profundidad muy baja puede llevar a un modelo subajustado que no capture las relaciones; mientras que una profundidad muy alta puede llevar a un sobreajuste.

```python
def objective_brf(trial):
    brf_params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 500),
        'max_depth': trial.suggest_int('max_depth', 3, 15),
        'max_features': 'sqrt',
        'sampling_strategy': 'all',
        'replacement': True,
        'bootstrap': False
        
    }
    
    pipeline_brf = Pipeline([
        ('Preprocessing', col_transformer),
        ('BalancedRF', BalancedRandomForestClassifier(**brf_params, random_state=555))
    ])
    
    pipeline_brf.fit(X_train, y_train)
    y_pred = pipeline_brf.predict(X_test)
    
    class_report = classification_report(y_test, y_pred, output_dict = True)
    
    f1_mob = class_report['1']['f1-score']
    
    return f1_mob
```

#### 3.1.3 Cálculo de Jensen-Shannon Divergence

A continuación, se utiliza *Jensen-Shannon Divergence* para medir si las nuevas *features* incorporadas tinen la misma distribución que las utilizadas durante la primera iteración. Para esto, se dividen según al tipo que corresponden, como se observó en la sección de descripción de los datos (2.1.2). Adicionalmente, se considera como diferencia notoria entre las distribuciones si el valor de JSD es mayor o igual a 0.5, considerando que esa *feature* presenta *data-drift*.

```python
categorical_features = ['JobStatus', 'AliveSession', 'CribStatus', 'LootMethod', 'InfoSource','DeviceOS']
categorical_numerical_features = ['DeviceEmails8w', 'HustleMonth', 'FreeMail', 'HomePhoneCheck', 'ForeignHustle', 'DeviceScams', 'CellPhoneCheck', 'ExtraPlastic']
numeric_features = ['DaysSinceJob', 'CreditCap', 'Speed24h', 'BankSpots8w', 'HustleMinutes', 'RiskScore', 'AliasMatch', 'ZipHustle',
                    'Speed4w', 'income', 'BankMonths', 'DOBEmails4w', 'OldHoodMonths', 'intended_balcon_amount', 'NewCribMonths', 'Speed6h',
                    'customer_age']
```

```python
def calculate_jsd(prob_dist_1, prob_dist_2):
    return jensenshannon(prob_dist_1, prob_dist_2)**2
```

```python
def preprocess_categorical(data, column):
    if pd.api.types.is_numeric_dtype(data[column]):
        value_counts = data[column].value_counts(normalize=True)
        return value_counts.sort_index().values
    else:
        value_counts = data[column].value_counts(normalize=True)
        return value_counts.sort_index().values
```

```python
def preprocess_continuous(data, column, bins=10):
    hist, bin_edges = np.histogram(data[column], bins=bins, density=True)
    hist = hist / hist.sum()
    return hist
```

```python
for i in categorical_features:
    cat_dist_1 = preprocess_categorical(X, i)
    cat_dist_2 = preprocess_categorical(X_new, i)
    try:
        jsd_cat = calculate_jsd(cat_dist_1, cat_dist_2)
        if jsd_cat < 0.5:
            print(f'JDS {i}', jsd_cat)
        else:
            print(f'----- > JDS {i}', jsd_cat, 'Mayor o igual a 0.5.')
            
    except:
        print(f'----- > JDS {i}', 'No se puede calcular.')
        
for i in categorical_numerical_features:
    cat_dist_1 = preprocess_categorical(X, i)
    cat_dist_2 = preprocess_categorical(X_new, i)
    try:
        jsd_cat = calculate_jsd(cat_dist_1, cat_dist_2)
        if jsd_cat < 0.5:
            print(f'JDS {i}', jsd_cat)
        else:
            print(f'----- > JDS {i}', jsd_cat, 'Mayor o igual a 0.5.')
    except:
        print(f'----- >JDS {i}', 'No se puede calcular.')
        
for i in numeric_features:
    cat_dist_1 = preprocess_continuous(X, i)
    cat_dist_2 = preprocess_continuous(X_new, i)
    try:
        jsd_cat = calculate_jsd(cat_dist_1, cat_dist_2)
        if jsd_cat < 0.5:
            print(f'JDS {i}', jsd_cat)
        else:
            print(f'----- > JDS {i}', jsd_cat, 'Mayor o igual a 0.5.')
    except:
        print(f'----- >JDS {i}', 'No se puede calcular.')
        
i = 'is_mob'
cat_dist_1 = preprocess_continuous(pd.DataFrame(y), i)
cat_dist_2 = preprocess_continuous(pd.DataFrame(y_new), i)
try:
    jsd_cat = calculate_jsd(cat_dist_1, cat_dist_2)
    if jsd_cat < 0.5:
            print(f'JDS {i}', jsd_cat)
    else:
        print(f'----- > JDS {i}', jsd_cat, 'Mayor o igual a 0.5.')
except:
    print(f'----- >JDS {i}', 'No se puede calcular.')
```

### 3.2 Intregación de Trackeo en el Modelamiento e interpretabilidad con MLFlow

Para el trackeo del modelamiento con MLFlow, se registran las métricas de *accuracy* global del modelo y *precision*, *recall* y *f1-score* de la clase fraudulenta. Adicionalmente, se guardan los gráficos de estas métricas por iteración, con tal de monitorear el avance del rendimiento de la optimización, y el *pipeline* resultante con el mejor modelo, que se usará posteriormente para el despliegue. 

```python
# Set de entrenamiento y validacion
X_train_new1, X_test_new, y_train_new1, y_test_new = train_test_split(X_new, y_new, test_size=0.1, stratify = y_new, random_state=555)

# Separar set de entrenamiento en dos
X_train_new1, X_train_new2, y_train_new1, y_train_new2 = train_test_split(X_train_new1, y_train_new1, test_size=0.5, stratify = y_train_new1, random_state=555)
```

```python
def plot_metrics(study, metric_name, plot_dir):
    values = [trial.values[0] for trial in study.trials]
    plt.figure()
    plt.plot(values)
    plt.title(f'{metric_name} over trials')
    plt.xlabel('Trial')
    plt.ylabel(metric_name)
    metric_plot_path = f"{plot_dir}/{metric_name}_history.png"
    plt.savefig(metric_plot_path)
    plt.close()
    mlflow.log_artifact(metric_plot_path, artifact_path="plots")
```

```python
def objective_brf_flow(trial):
    features_selected = ['MinMaxScaler__CreditCap', 'MinMaxScaler__BankSpots8w', 'MinMaxScaler__HustleMinutes', 'MinMaxScaler__RiskScore', 
                         'MinMaxScaler__ZipHustle', 'MinMaxScaler__Speed4w', 'MinMaxScaler__BankMonths', 'MinMaxScaler__DOBEmails4w', 'MinMaxScaler__intended_balcon_amount', 
                         'MinMaxScaler__NewCribMonths', 'MinMaxScaler__Speed6h', 'MinMaxScaler__customer_age', 'StandardScaler__Speed24h', 'RobustScaler__DaysSinceJob', 
                         'RobustScaler__OldHoodMonths', 'OneHot__JobStatus_CA', 'OneHot__JobStatus_CB', 'OneHot__CribStatus_BA', 'OneHot__CribStatus_BB', 
                         'OneHot__CribStatus_BC', 'OneHot__CribStatus_BE', 'OneHot__LootMethod_AA', 'OneHot__LootMethod_AB', 'OneHot__LootMethod_AC',
                         'OneHot__DeviceOS_linux', 'OneHot__DeviceOS_other', 'OneHot__DeviceOS_windows', 'Pasthrough__AliveSession', 'Pasthrough__AliasMatch', 
                         'Pasthrough__DeviceEmails8w', 'Pasthrough__income', 'Pasthrough__FreeMail', 'Pasthrough__HomePhoneCheck', 'Pasthrough__ExtraPlastic']
    
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 200, 1500),
        'max_features': 'sqrt',
        'sampling_strategy': 'all',
        'replacement': True,
        'bootstrap': False
    }
        
    select_columns_transformer = FunctionTransformer(lambda X: X[features_selected])

    pipeline_brf = Pipeline([('Preprocessing', col_transformer), 
                             ('SelectColumns', select_columns_transformer),
                             ('BalancedRFC', BalancedRandomForestClassifier(**params, random_state=555, warm_start=True))
                            ])
    
    pipeline_brf.set_output(transform='pandas')
    
    pipeline_brf.fit(X, y)
    pipeline_brf['BalancedRFC'].set_params(n_estimators=params['n_estimators']*2)
    pipeline_brf.fit(X_train_new1, y_train_new1)
    pipeline_brf['BalancedRFC'].set_params(n_estimators=params['n_estimators']*3)
    pipeline_brf.fit(X_train_new2, y_train_new2)
    
    y_pred = pipeline_brf.predict(X_test_new)
    class_report = classification_report(y_test_new, y_pred, output_dict=True)

    f1_mob = class_report['1']['f1-score']
    recall_mob = class_report['1']['recall']
    precision_mob = class_report['1']['precision']
    global_accura = class_report['accuracy']
    
    # Registrar los resultados de este trial en MLFlow
    with mlflow.start_run(nested=True, run_name=f"BRF with n_estimators: {params['n_estimators']}"):
        mlflow.log_params(params)
        mlflow.log_metric("f1-score_fraudulent", f1_mob)
        mlflow.log_metric("precision_fraudulent", precision_mob)
        mlflow.log_metric("recall_fraudulent", recall_mob)
        mlflow.log_metric("global_accuracy", global_accura)
        mlflow.sklearn.log_model(pipeline_brf['BalancedRFC'], "model")

    return f1_mob
```

```python
def get_best_pipeline(experiment_id):
    runs = mlflow.search_runs(experiment_id)
    best_run_id = runs.sort_values("metrics.f1-score_fraudulent", ascending=False)["run_id"].iloc[0]
    best_pipeline = mlflow.pyfunc.load_model("runs:/" + best_run_id + "/model")
    return best_pipeline
```

```python
def retrain_optimize():
    remote_server_uri = 'http://127.0.0.1:8080'
    mlflow.set_tracking_uri(remote_server_uri)
    experiment_name = "mlflow-retrain-optimize"
    mlflow.set_experiment(experiment_name)

    # Define un nombre único para esta ejecución
    run_name = "Optimization and Retraining"

    with mlflow.start_run(run_name=run_name):
        sampler = optuna.samplers.TPESampler(seed=555)
        study = optuna.create_study(direction='maximize', sampler=sampler)
        study.optimize(objective_brf_flow, n_trials=50)
        
        # Guardar gráficos de Optuna
        plot_dir = "plots"
        os.makedirs(plot_dir, exist_ok=True)

        plot_optimization_history = optuna.visualization.plot_optimization_history(study)
        plot_optimization_history.write_image(f"{plot_dir}/optimization_history.png")
        mlflow.log_artifact(f"{plot_dir}/optimization_history.png", artifact_path="plots")

        plot_metrics(study, 'f1-score_fraudulent', plot_dir)
        plot_metrics(study, 'precision_fraudulent', plot_dir)
        plot_metrics(study, 'recall_fraudulent', plot_dir)
        plot_metrics(study, 'global_accuracy', plot_dir)

        best_pipeline = get_best_pipeline(mlflow.get_experiment_by_name(experiment_name).experiment_id)
        pipeline_dir = "models"
        os.makedirs(pipeline_dir, exist_ok=True)
        with open(f"{pipeline_dir}/brf_best_pipeline.pkl", 'wb') as f:
            pickle.dump(best_pipeline, f)
        mlflow.log_artifact(f"{pipeline_dir}/brf_best_pipeline.pkl", artifact_path="models")

    mlflow.end_run()
```

A lo largo del proyecto, se iteró el modelo seleccionado 3 veces, una por fase establecida, utilizando los nuevos datos proporcionados y reentrenando y optimizando el modelo. Los ajustes realizados se muestran a continuación:

Luego, se estudio la mejor combinación que aumentara el rendimiento de este, por medio de la la eliminación de *features*, con *Recursive Feature Elimination*, y optimización de hiperparámetros utilizando Optuna.

2. **Iteración 2**

Al incorporar una mayor cantidad de datos, se pide realizar *partial fit* para el entrenamiento del modelo. Como BRF no posee *partial fit* (cada *fit* que se haga reentrena un modelo con los datos entrengados), se utiliza el hiperparámetro *warm_start*, que guarda los estimadores generados en el *fit* anterior y entrena un nuevo set de estos. De esta manera, se resuelve parcialmente el problema de *partial fit*, a costa de la generación de un modelo más robusto pero más costoso en términos computacionales (pesa mucho). Adicionalmente, se llevó a cabo un nuevo proceso de optimización de hiperparámetros, variando la cantidad de estimadores del modelo.

3. **Iteración 3**

Similar a lo realizado en la segunda iteración. Se incorporan nuevos datos y se reentrena el modelo por medio del *partial fit* propuesto, además de la optimización de hiperparámetros.


Los resultados de estos se muestran en términos de AUC-PR, que corresponde al Área Bajo la Curba de *Precision-Recall*. Esta métrica de evaluación se basa en la relación entre la tasa de verdaderos positivos y la tasa de falsos positivos, centrandose en la precisión y la sensibilidad del modelo. El beneficio de esta métrica se basa en que proporciona una mejor compresión del desempeño del modelo con respecto a la clase minoria, que es más relevante en este contexto donde poseemos datos desbalanceados y se busca identificar a la clase con menor cantidad de muestras (fraudulentos).

Esto se puede deber a distintos motivos, teniendo como principales hipótesis que los datos entre iteración son muy variables, por lo que el modelo se subajusta y no encuentra las relaciones o patrones correctamente, teniendo que aumentar la cantidad de estimadores para que se establezcan reglas que aborden de mejor manera los casos expuesto; se tienen *features* poco relevantes para el modelo que generan ruido y, por consiguiente, un peor desempeño, por lo que se debería realizar una evaluación de estas de manera más minuciosa y realizar todo el proceso de modelamiento nuevamente, bajo el supuesto de ''*garbage in - gargabe out*.''
